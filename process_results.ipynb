```python
import numpy as np
from scipy.io import loadmat, savemat
import tensorflow as tf
```

```python
datadict = loadmat('split_set.mat')
val_imgs = datadict['val_imgs']
val_target = datadict['val_target'][0]
```

```python
All_loss = np.zeros([60, 50])
All_accuracy = np.zeros([60, 50])
All_val_loss = np.zeros([60, 50])
All_val_accuracy = np.zeros([60, 50])
All_results = np.zeros([60, 2])
All_confusion = np.zeros([60, 3, 3])

for i in range(60):
    model = tf.keras.models.load_model(f'Test_models/model{i+1}.hdf5')
    results = np.array(model.evaluate(val_imgs, val_target))
    predictions = np.argmax(model.predict(val_imgs), axis=1)
    confusion_matrix = np.array(tf.math.confusion_matrix(val_target, predictions))
    histdict = loadmat(f'Test_models/model{i+1}.mat')

    All_loss[i, :] = histdict['loss'][0, :]
    All_accuracy[i, :] = histdict['accuracy'][0, :]
    All_val_loss[i, :] = histdict['val_loss'][0, :]
    All_val_accuracy[i, :] = histdict['val_accuracy'][0, :]
    All_results[i, :] = results
    All_confusion[i, :, :] = confusion_matrix

outdict = {}
outdict['All_loss'] = All_loss
outdict['All_accuracy'] = All_accuracy
outdict['All_val_loss'] = All_val_loss
outdict['All_val_accuracy'] = All_val_accuracy
outdict['All_results'] = All_results
outdict['All_confusion'] = All_confusion
savemat('results.mat', outdict)
```
