```python
import numpy as np
from scipy.io import loadmat
```

```python
datadict = loadmat('results.mat')
args = np.argmax(datadict['All_val_accuracy'], axis = 1)
```

```python
print('loss')
values = datadict['All_loss'][np.arange(60), args]
average = np.average(values)
standard_deviation = (1/59 * np.sum((values-average)**2))**0.5
print('Average = ', average)
print('Standard deviation = ', standard_deviation)
```

```python
print('accuracy')
values = datadict['All_accuracy'][np.arange(60), args]
average = np.average(values)
standard_deviation = (1/59 * np.sum((values-average)**2))**0.5
print('Average = ', average)
print('Standard deviation = ', standard_deviation)
```

```python
print('val_loss')
values = datadict['All_val_loss'][np.arange(60), args]
average = np.average(values)
standard_deviation = (1/59 * np.sum((values-average)**2))**0.5
print('Average = ', average)
print('Standard deviation = ', standard_deviation)
```

```python
print('confusion_matrix')
values = datadict['All_confusion']
average = np.average(values, axis=0)
standard_deviation = (1/59*np.sum((values - np.tile(average, (60, 1, 1)))**2, axis = 0))**0.5

average = average[:, [1, 2, 0]]; average = average[[1, 2, 0], :]
standard_deviation = standard_deviation[:, [1, 2, 0]]; standard_deviation = standard_deviation[[1, 2, 0], :]

print('Average = \n', average)
print('Standard deviation = \n', standard_deviation)
```
